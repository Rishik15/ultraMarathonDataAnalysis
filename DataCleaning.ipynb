```python
import pandas as pd
```

```python
marathon_df = pd.read_csv('TWO_CENTURIES_OF_UM_RACES.csv')
marathon_df.head()
```

```python
marathon_df.shape
```

```python
marathon_df.info()
```

#### Seperating only data with Marathon distance as 100km and taken place in IND or USA in the year 2021.

```python
marathon_df['Event distance/length'].value_counts()
```

```python
marathon_df[(marathon_df['Event distance/length'].isin(['100km'])) & (marathon_df['Year of event'] == 2021)]
```

```python
condition = marathon_df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0).isin(['USA','IND'])
```

```python
M2021 = marathon_df[(marathon_df['Event distance/length'].isin(['100km'])) & (marathon_df['Year of event'] == 2021) & condition]
```

```python
M2021.shape
```

```python
M2021.info()
```

#### Removing the null values in age category

```python
M2021.isnull().sum()
```

```python
M2021 = M2021.dropna(subset='Athlete year of birth')
```

```python
M2021.shape
```

#### Removing the 'h' from Athlete performance and converting the column data type into a timedelta object

```python
M2021['Athlete performance'] = M2021['Athlete performance'].str.split(' ').str.get(0)
```

```python
M2021.head()
```

```python
M2021['Athlete performance'] = pd.to_timedelta(M2021['Athlete performance'])
M2021.head()
```

#### Creating a new columnn named 'Athlete_age' with the ages of all the Athletes during 2021

```python
M2021['Athlete_age'] = (2021 - M2021['Athlete year of birth']).astype(int)
```

```python
M2021.head()
```

```python
M2021.info()
```

#### Dropping columns:'Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'

```python
M2021 = M2021.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis = 1)
```

```python
M2021 = M2021.reset_index(drop = True)
```

```python
M2021.isna().sum()
```

#### Checking for any duplicates

```python
M2021.duplicated().sum()
```

#### Converting the rest of the columns into correct data types

```python
M2021.info()
```

```python
M2021['Athlete average speed'] = M2021['Athlete average speed'].astype(float)
```

```python
M2021.info()
```

```python
M2021.head()
```

#### Renaming column names to proper conventional names for easy handling

```python
M2021 = M2021.rename(columns={'Year of event':'year',
                              'Event dates':'event_date',
                              'Event name':'event_name',
                              'Event distance/length':'distance',
                              'Event number of finishers':'number_of_finishers',
                              'Athlete performance':'athlete_performance',  
                              'Athlete gender':'athlete_gender',         
                              'Athlete average speed':'athlete_average_speed',
                              'Athlete ID':'athlete_id',
                              'Athlete_age':'athlete_age',
}   
)
```

```python
M2021.head()
```

#### Re-ordering the columns without the year column

```python
M2021 = M2021[['event_date','event_name','distance','number_of_finishers','athlete_id','athlete_gender','athlete_age','athlete_performance','athlete_average_speed']]
```

```python
M2021
```

```python
M2021.to_csv('cleaned_M2021.csv', index = True)
```
